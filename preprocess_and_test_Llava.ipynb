```python
def is_too_white(img, threshold=252.5):
    img_array = np.array(img)
    mean_value = np.mean(img_array)
    return mean_value > threshold

def process_images(folder_path):
    valid_files = []

    for filename in sorted(os.listdir(folder_path)):
        if filename.lower().endswith('.png'):
            file_path = os.path.join(folder_path, filename)

            try:
                with Image.open(file_path) as img:
                    width, height = img.size

                    if width < 85 or height < 40:
                        os.remove(file_path)
                        print(f"Removed {filename} - too small")
                    elif is_too_white(img):
                        os.remove(file_path)
                        print(f"Removed {filename} - too white")
                    else:
                        valid_files.append(filename)
                        print(f"Kept {filename}")
            except Exception as e:
                print(f"Error processing {filename}: {e}")

    return valid_files
```

```python
def process_files(folder_path, doc_name):
    results = {}

    png_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.png')]

    for png_file in png_files:
        base_name = os.path.splitext(png_file)[0].lower()

        md_file = doc_name + '.md'
        md_path = os.path.join(folder_path, md_file)

        if os.path.exists(md_path):
            with open(md_path, 'r', encoding='utf-8') as file:
                content = file.read().lower()

                match = re.search(r'\b' + re.escape(base_name) + r'\b', content)

                if match:
                    start = max(0, match.start() - 350)
                    excerpt = content[start:match.start()]

                    if len(excerpt) < 350:
                        excerpt = content[:350]

                    results[png_file] = excerpt.strip()
                else:
                    results[png_file] = "Image name not found in the corresponding .md file"
        else:
            results[png_file] = "Corresponding .md file not found"

    return results


folder_path = '/content/drive/MyDrive/AI_Talent_Hub/formated_markdown/doc_1'
file_excerpts = process_files(folder_path, 'doc_1')
file_excerpts
```

```python
def extract_after_assistant(text):
    split_text = text.split("АССИСТЕНТ", 1)
    if len(split_text) > 1:
        return split_text[1].strip()
    else:
        return ""
```

```python
def process_images_and_update_md(input_directory, output_directory, doc_name, start):

    os.makedirs(output_directory, exist_ok=True)

    png_files = [f.lower() for f in os.listdir(input_directory) if f.lower().endswith('.png')]

    doc_path = os.path.join(input_directory, f'{doc_name}.md')

    if not os.path.exists(doc_path):
        print("Файл doc.md не найден в указанной директории.")
        return

    with open(doc_path, 'r', encoding='utf-8') as file:
        content = file.read().lower()

    file_mapping = {}

    for i, png_file in enumerate(png_files, start=start):
        old_name = png_file
        new_name = f'image_{i}.png'

        pattern = r'!\[' + re.escape(old_name) + r'\]\(' + re.escape(old_name) + r'\)'
        replacement = f'$image_{i}$'

        content = re.sub(pattern, replacement, content)
        file_mapping[old_name] = new_name

    with open(doc_path, 'w', encoding='utf-8') as file:
        file.write(content)

    for old_name, new_name in file_mapping.items():
        old_path = os.path.join(input_directory, old_name)
        new_path = os.path.join(output_directory, new_name)

        for filename in os.listdir(input_directory):
            if filename.lower() == old_name:
                old_path = os.path.join(input_directory, filename)
                break

        shutil.copy2(old_path, new_path)
        print(f"Скопирован и переименован файл: {old_name} -> {new_name}")

    print(f"Обработано {len(png_files)} файлов. Файл doc.md обновлен.")
    if png_files:
        return i
    else:
        return start
```

```python
def clean_md_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    lowercase_content = content.lower()

    pattern = r'!\[\d+_image_\d+\.png\]\(\d+_image_\d+\.png\)'

    matches = list(re.finditer(pattern, lowercase_content))

    for match in reversed(matches):
        start, end = match.span()
        content = content[:start] + content[end:]

    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(content)

    print(f"Файл {file_path} обработан. Ссылки на изображения удалены.")
```
